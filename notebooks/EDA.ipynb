```python
_DS_PATH = Path("../dataset")
train_ds = spcup23_ds(_DS_PATH, test=False)
test_ds = spcup23_ds(_DS_PATH, test=True)

len(train_ds), len(test_ds)
```

```python
for i, (filename, label, fnc, icn_tc) in enumerate(train_ds):

    print("Folder:\t\t", filename)
    print("Etykieta:\t", label)
    print("fnc.shape:\t", fnc.shape)
    print("icn_tc.shape:\t", icn_tc.shape)

    fig, (ax1,ax2,ax3) = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
    ax1.set_title("Cała macierz korelacji")
    ax1.imshow(icn_tc.T @ icn_tc)

    ax2.set_title("Sygnał wielokanałowy")
    ax2.imshow(icn_tc.T)

    for k in range(105):
        ax3.plot(icn_tc[:, k])
    plt.show()

    if i > 2: break
```
